```python
import numpy as np
import matplotlib.pyplot as plt
import random
import cv2
import PIL
from PIL import Image
```

```python
#for rectangle
```

```python
path = r'C:\Users\ganap\Downloads\A_black_background (1).jpg'
image = cv2.imread(path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

for i in range(0,399):
    r=random.randint(100,600)
    f=random.randint(300,800)
    
    start_point = (r, f)
    end_point = (220, 220)
    color = (255, 255, 255)
    thickness = -5
  
    image = cv2.rectangle(image, start_point, end_point, color, thickness)
    plt.figure(figsize=(20, 20))
    plt.subplot(2, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    data1 = np.array(image)
    image = cv2.imread(path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
```

```python
#for circle
```

```python
for i in range(0,399):
    r=random.randint(300,800)
    f=random.randint(300,900)
    c=random.randint(30,500)
   
    center_coordinates = (r, f)
 

    radius = c

    color = (255, 255, 255)
  
    thickness = -12
      
    image = cv2.circle(image, center_coordinates, radius, color, thickness)
  
   
    plt.figure(figsize=(20, 20))
    plt.subplot(2, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    data2 = np.array(image)
    image = cv2.imread(path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```
